# Data analysis

```python
#Average rating
foc_df['Rating'].mean()
```

```python
#Average rating by date
foc_df.groupby('Review_Date').mean()['Rating']
```

```python
#Cocoa percentage by country

foc_df.groupby('Company_Location').Cocoa_Percent.mean()
```

```python
foc_df['Bean_Type'].fillna('MISSING', inplace=True)
foc_df.head()
```

```python
# Top 5 companies by rating
foc_df.sort_values(by = 'Rating', ascending = False).head(5)
```

```python
# Bottom 5 companies by rating
foc_df.sort_values(by = 'Rating', ascending = False).tail(5)
```

```python
# Companies in Lithuania
foc_df[foc_df['Company_Location']=='Lithuania']
```

```python
# Average ratings by country 
foc_df.groupby('Company_Location').mean()['Rating'].sort_values()
```

```python
# Top 100 companies by rating
foc_df.sort_values(by = 'Rating', ascending = False).head(100)
```

```python
top100 = foc_df.sort_values(by = 'Rating', ascending = False).head(100)
top100['Company'].value_counts()
```

# Data vizualization

```python
# Average rating by year
rating_year = foc_df.groupby('Review_Date').mean()['Rating']
rating_year.plot()
```

```python
rating_year = foc_df.groupby('Review_Date').mean()['Cocoa_Percent']
rating_year.plot()
```

```python
# Correlations
foc_df['Rating'].corr(foc_df['Cocoa_Percent'])
```
